```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("/kaggle/input/homeprice/homeprices.csv")
```

```python
df
```

# Data preprocessing : Fill NAN values with median value of a column

```python
df.bedrooms.median()
```

```python
df.bedrooms = df.bedrooms.fillna(df.bedrooms.median())
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df.drop('price', axis='columns'), df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

# Find price of home with 3000 sqr ft area, 3 bedrooms , 40 year old

```python
reg.predict([[3000, 3, 40]])
```
